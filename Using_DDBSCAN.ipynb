```python
# Importar função para gerar dados aglomerados sintéticos
from sklearn.datasets.samples_generator import make_blobs
# Importar pacote gráfico
import matplotlib.pyplot as plt
# Desenhar gráficos no corpo do notebook
%matplotlib inline

# Criar três aglomerados de pontos bi-dimensionais
data, clusters = make_blobs(n_samples=300, centers=[[0,0], [10,15], [30,30]], cluster_std=3, n_features=2,
                     random_state=0)

# Extrair coordenadas x e y dos pontos gerados
x = zip(*data)[0]
y = zip(*data)[1]

# Cores a serem utilizadas para os aglomerados
colors = 'brgcmykw' * 100
```

```python
# Criar gráfico dos pontos coloridos de acordo com seu aglomerado
for index, cluster in enumerate(clusters):
    plt.scatter(x[index],y[index], color=colors[cluster])
plt.title('Aglomerados gerados')
```

```python
# Criar gráfico dos pontos com mesma cor
plt.scatter(x,y)
plt.title(u'Aglomerados encontrados na prática')
```

```python
# Acrescentar ruído
x = x + (20, 0, 15, 35, 40)
y = y + (0, 25, 2, 10, 5)
plt.scatter(x,y)
plt.title(u'Aglomerados encontrados na prática (com ruído)')
```

```python
import ddbscan

# Criar modelo buscando pontos em um raio de 5 unidades
# e requerendo no mínimo 25 pontos vizinhos para estar
# no centro de um aglomerado
model = ddbscan.DDBSCAN(eps=3, min_pts=10)

# Adicionar pontos ao modelo incrementalmente
for point in zip(x, y):
    model.add_point(point=list(point), count=1, desc="Just a point")
    
# Computar aglomerados
model.compute()

print 'Quantidade de aglomerados encontrados:', len(model.clusters)

# Criar gráfico com os aglomerados encontrados
for i in range(len(model.points)):
    cluster = model.points_data[i].cluster
    if cluster == -1: # Ruído
        plt.scatter(x[i],y[i], color='k', marker='x', s=40)
    else: # Aglomerado
        plt.scatter(x[i],y[i], color=colors[cluster])
```

```python
# Modificar parâmetros
model.set_params(eps=2, min_pts=5)

# Computar aglomerados
model.compute()

print 'Quantidade de aglomerados encontrados:', len(model.clusters)

# Criar gráfico com os aglomerados encontrados
for i in range(len(model.points)):
    cluster = model.points_data[i].cluster
    if cluster == -1: # Ruído
        plt.scatter(x[i],y[i], color='k', marker='x', s=40)
    else: # Aglomerado
        plt.scatter(x[i],y[i], color=colors[cluster])
```
